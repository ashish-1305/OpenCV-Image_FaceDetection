```python
import cv2
```

```python
model=cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
faces=model.detectMultiScale(face)
```

```python
faces
```

```python
faces.shape
```

```python
for face1 in faces:
    x,y,w,h=face1
    picture=cv2.rectangle(face,(x,y),(x+w,y+h),(0,0,0),1)
```

```python
plt.imshow(picture)
```
